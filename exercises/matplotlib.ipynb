```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
x = np.linspace(0, 10, 100)

#first subplot
ax_0 = plt.subplot(2, 1, 1) #2 rows, 1 column, index 1
ax_1 = plt.subplot(2, 1, 2) #2 rows, 1 column, index 1

x = np.arange(0, 10, 0.1)
y_0 = x ** 2
y_1 = np.cos(x)

ax_0.plot(x, y_0)
ax_1.plot(x, y_1)
```

```python
#USE THIS ONE BY DEFAULT, EASIER
fig, axes = plt.subplots(2, 2, figsize = (5,5))
axes[0, 1].plot(x, y_0)
```

### MATPLOTLIB GRAPHICS

```python
ax = plt.subplot()

x = np.arange(-5, 5, 0.1)
y = x **2

ax.plot(x, y_0, "r--")

ax.plot(x, y_1, "b-+")

ax.legend(["y = x^2", "y = cos(x)"])

ax.set_title("This represents: y = x^2")

ax.set_xlabel("x (s)")

ax.set_ylabel("y (m)")
```
